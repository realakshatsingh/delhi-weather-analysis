```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('kaggel_weather_2013_to_2024.csv')
df.head()
df.drop(columns='Unnamed: 0',inplace=True)
df.head()
```

```python
df.columns
```

```python
df= df[['DATE','temp','precip','tempmax','tempmin']]
print(df.shape)
print(df.isna().sum())
print("Data is clean and ready to work with")
df
```

```python
arr_date = df['DATE'].to_numpy()
arr_temp = df['temp'].to_numpy()
arr_temp_max = df['tempmax'].to_numpy()
arr_temp_min = df['tempmin'].to_numpy()
arr_prepcip = df['precip'].to_numpy()
```

```python
print("Phase 2")
print("Average Temperature",np.mean(arr_temp))
print("Maximum Temperature",np.max(arr_temp_max))
print("Minimum Temperature",np.min(arr_temp_min))
print("Standard Devaition of the Temperature",np.std(arr_temp))
print("Median Temperature",np.median(arr_temp))
```

```python
print("Phase 3 Temperature Analysis")
k = np.argmax(arr_temp_max)
print("Hottest Day",arr_date[k])
k = np.argmin(arr_temp_min)
print("Coldest Day",arr_date[k])

total = arr_date.shape[0]

no_above35 = np.sum(np.where(arr_temp_max>35,True,False))
percent_no_above35 = round((no_above35/total*100),2)
print("Days above 35 degree Celsius",percent_no_above35,"%")


no_below10 = np.sum(np.where(arr_temp_min<10,True,False))
percent_no_below10 = round((no_below10/total*100),2)
print("Days below 10 degree Celsius",percent_no_below10,"%")

list_of_days = np.argsort(arr_temp_max)
all_indexes = list_of_days[-1:-11:-1]
j = 1
for i in all_indexes:
    print(j,"Hottest Day",arr_date[i],arr_temp_max[i])
    j+=1
```
