```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_main = pd.read_csv('kaggel_weather_2013_to_2024.csv')
df_main.head()
df_main.drop(columns='Unnamed: 0',inplace=True)
df_main.head()
```

```python
df_main.columns
```

```python
df = df_main[['DATE','temp','precip','tempmax','tempmin']]
print(df.shape)
print(df.isna().sum())
print("Data is clean and ready to work with")
df
```

```python
arr_date = df['DATE'].to_numpy()
arr_temp = df['temp'].to_numpy()
arr_temp_max = df['tempmax'].to_numpy()
arr_temp_min = df['tempmin'].to_numpy()
arr_prepcip = df['precip'].to_numpy()
```

```python
print("Phase 2")
print("Average Temperature",np.mean(arr_temp))
print("Maximum Temperature",np.max(arr_temp_max))
print("Minimum Temperature",np.min(arr_temp_min))
print("Standard Devaition of the Temperature",np.std(arr_temp))
print("Median Temperature",np.median(arr_temp))
```

```python
print("Phase 3 Temperature Analysis")
k = np.argmax(arr_temp_max)
print("Hottest Day",arr_date[k])
k = np.argmin(arr_temp_min)
print("Coldest Day",arr_date[k])

total = arr_date.shape[0]

no_above35 = np.sum(np.where(arr_temp_max>35,True,False))
percent_no_above35 = round((no_above35/total*100),2)
print("Days above 35 degree Celsius",percent_no_above35,"%")


no_below10 = np.sum(np.where(arr_temp_min<10,True,False))
percent_no_below10 = round((no_below10/total*100),2)
print("Days below 10 degree Celsius",percent_no_below10,"%")

list_of_days = np.argsort(arr_temp_max)
all_indexes = list_of_days[-1:-11:-1]
j = 1
for i in all_indexes:
    print(j,"Hottest Day",arr_date[i],arr_temp_max[i])
    j+=1
```

```python
indexes_for_rain = df_main['DATE'].str.contains("2023")
arr_rain = df_main['precip'].to_numpy()
rain_measure = np.sum(arr_rain[indexes_for_rain])
print(rain_measure,"mm rainfall occured in year 2023")
```

```python
avg_rainfall = round(np.mean(arr_rain),2)
print("The average rainfall is",avg_rainfall,"mm/day in Delhi over the years from April 2013 to November 2024")
```

```python
max_rainfall = np.max(arr_rain)
max_rainfall_day = arr_date[np.argmax(arr_rain)]
print(f"The maximum rainfall {max_rainfall}mm, occured on {max_rainfall_day}")
```

```python
rainfall_days = np.where(arr_rain>0,True,False).sum()
total_days = total
k = rainfall_days/total_days*100
print("Percent Days with rain",k,"%")
print("Percent Days without rain",100-k,"%")
```

Phase 5 Monthly Analysis

```python
def month_extractor(dateinstr):
    list_of_date = dateinstr.split("/")
    return list_of_date[0]


month_getter = np.vectorize(month_extractor)
arr_month = month_getter(arr_date)
arr_month
```

```python
temp_data_holder = {}
for i in range(1,13,1):
    print("For",i,"Month")
    indices = np.where(month_getter(arr_date)==str(i),True,False)
    print("Average Temp of Month",i,"is",end=" ")
    print(np.mean(arr_temp[indices]),"Celsius")
    temp_data_holder[i] = np.mean(arr_temp[indices])
    
```

```python
max = arr_temp[0]
month = 1
for i in range(1,13,1):
    indices = np.where(month_getter(arr_date)==str(i),True,False)
    temp_for_month = np.mean(arr_temp_max[indices])
    if max < temp_for_month:
        max = temp_for_month
        month = i
print("The Hottest Month is",month)

    
    
    
```

```python
def year_str_getter(dateinstr):
    list_of_date = dateinstr.split("/")
    return list_of_date[-1]
year_getter = np.vectorize(year_str_getter)
data = {}
for i in range(2013,2025,1):
    for j in range(1,13,1):
        indices = (month_getter(arr_date)==str(j)) & (year_getter(arr_date)==str(i)) # and operation
        if np.sum(indices) == 0:
            continue

        if j in data:
            rain_month = np.sum(arr_prepcip[indices])
            data[j].append(rain_month)
            
        else:
            rain_month = np.sum(arr_prepcip[indices])
            data[j] = [rain_month]
            
data_final = {}
for month in range(1,13,1):
    lst_for_month = data[month]
    agg = 0
    for j in lst_for_month:
        agg+=j
    mean = agg/len(lst_for_month)
    print("For month",month,"the average rain is",mean,"mm")
    data_final[month] = mean

max = data_final[1]
month = 1
for i in data_final:
    if data_final[i]>max:
        max = data_final[i]
        month = i
print("The wettest month is",month)

        

```
